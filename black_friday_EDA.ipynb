# Black Friday Data Cleaning and Feature Engineering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# PROBLEM STATEMENT #
A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
bf_train =pd.read_csv('/content/train.csv')
```

```python
bf_train.head()
```

```python
bf_test = pd.read_csv('/content/test.csv')
bf_test.head()
```

# Combining Test and train data for analysis

```python
bfdata = bf_train.append(bf_test)
```

```python
bfdata.info()
```

```python
bfdata.describe()
```

# Deleting useless Columns

```python
bfdata.drop(['User_ID'],axis=1,inplace=True)
```

# Changing Gender category to 1 and 0

```python
bfdata['Gender'] = bfdata['Gender'].map({'M':1,'F':0})
```

```python
bfdata
```

# Changing Age Feature for Feature Engineering

```python
bfdata['Age'].unique()
```

```python
bfdata['Age']=bfdata['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
bfdata
```

# Fixing City categorical data


```python
bfcity = pd.get_dummies(bfdata['City_Category'],drop_first=True)
```

```python
bfdata = pd.concat([bfdata,bfcity],axis=1)
```

```python
bfdata.drop(['City_Category'],axis=1,inplace=True)
```

```python
bfdata.head()
```

# Replacing Missing values

```python
bfdata.isnull().sum()
```

```python
bfdata['Product_Category_2'] = bfdata['Product_Category_2'].fillna( bfdata['Product_Category_2'].mode()[0])
```

```python
bfdata['Product_Category_3'] = bfdata['Product_Category_3'].fillna( bfdata['Product_Category_3'].mode()[0])
```

# Converting object into Integer

```python
bfdata['Stay_In_Current_City_Years']=bfdata['Stay_In_Current_City_Years'].str.replace('+',' ')
```

```python
bfdata['Stay_In_Current_City_Years']=bfdata['Stay_In_Current_City_Years'].astype(int)
```

```python
bfdata['B'] = bfdata['B'].astype(int)
bfdata['C'] = bfdata['C'].astype(int)
```
